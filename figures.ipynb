```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def pdf(x):
    a = 1/57
    return (a**2) * x * np.exp(-0.5 * (a**2) * (x**2))

def cdf(x):
    a = 1/57
    return 1 - (np.exp(-0.5 * (a**2) * (x**2)))

def invcdf(u):
    a = 1/57
    return np.sqrt(((-2)/(a**2)) * np.log(1-u))
```

```python
xRange = np.linspace(0, 225, num = 500)
yPDFRange = [pdf(x) for x in xRange]
yCDFRange = [cdf(x) for x in xRange]

plt.xlabel("Distance")
plt.ylabel("Probability Density")
plt.plot(xRange, yPDFRange)
plt.show()
```

```python
plt.xlabel("Distance")
plt.ylabel("Cumulative Probability")
plt.plot(xRange, yCDFRange)
plt.show()
```

```python
r1 = invcdf(0.5)
r2 = invcdf(0.7)
r3 = invcdf(0.9)
lim = 170

figure, axes = plt.subplots()
c1 = plt.Circle((0, 0), r1 , alpha=0.4, color = "green")
c2 = plt.Circle((0, 0), r2, alpha=0.2, color = "orange")
c3 = plt.Circle((0, 0), r3, alpha=0.1, color = "red")
axes.add_artist( c1 )
axes.add_artist( c2 )
axes.add_artist( c3 )
axes.set_aspect(1)
plt.title( 'Landing Zones' )
plt.xlim(-lim, lim)
plt.ylim(-lim, lim)
plt.show()
```
